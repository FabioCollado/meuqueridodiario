```python
def break_longer_paragraphs(text):
    tokenized_text = tokenizer.tokenize(text)
    if len(tokenizer.tokenize(preprocess2(text))) <= 510 or len(tokenized_text) <= 510:
        return [text]
    
    for i in range(509, len(tokenized_text), 10):
        if len(tokenizer.tokenize(preprocess2(tokenizer.convert_tokens_to_string(tokenized_text[:i])))) > 498:
            break
    truncated_text = [tokenizer.convert_tokens_to_string(tokenized_text[:i])]
    rest_of_text = break_longer_paragraphs(tokenizer.convert_tokens_to_string(tokenized_text[i:]))
    return  truncated_text + rest_of_text
```

```python
data2 = []
for i, file in enumerate(data):
    print(i,'/',len(data))
    diario = []
    for paragraph in file:
        diario = diario + break_longer_paragraphs(paragraph)
    data2.append(diario)
```

# Outras opções para converter PDF

```python
from io import StringIO
from pdfminer.pdfinterp import PDFResourceManager, PDFPageInterpreter
from pdfminer.converter import TextConverter
from pdfminer.layout import LAParams
from pdfminer.pdfpage import PDFPage
import os
import sys, getopt
#converts pdf, returns its text content as a string
def convert(file_name, pages=None):
    if not pages:
        pagenums = set()
    else:
        pagenums = set(pages)

    output = StringIO()
    manager = PDFResourceManager()
    converter = TextConverter(manager, output, laparams=LAParams())
    interpreter = PDFPageInterpreter(manager, converter)
    print(file_name)
    infile = open(file_name, 'rb')
    page_number = 0
    for page in PDFPage.get_pages(infile, pagenums):
        page_number+=1
        interpreter.process_page(page)
        if page_number == 3: break
    infile.close()
    converter.close()
    text = output.getvalue()
    output.close
    return text 

def pdfparser(data):

    fp = open(data, 'rb')
    manager = PDFResourceManager()
    retstr = StringIO()
    codec = 'utf-8'
    laparams = LAParams()
    device = TextConverter(manager, retstr, laparams=laparams)
    # Create a PDF interpreter object.
    interpreter = PDFPageInterpreter(manager, device)
    # Process each page contained in the document.

    for page in PDFPage.get_pages(fp):
        interpreter.process_page(page)
        data =  retstr.getvalue()
        print(data)
file_name = "./pdfs/008dbca0cbf58ecd9f468848e62d36e40fecfedf.pdf"
t = pdfparser(file_name)
t
```
